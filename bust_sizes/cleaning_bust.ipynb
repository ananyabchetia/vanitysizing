```python
import pandas as pd
df=pd.read_csv('csv/bust.csv')
df
```

## Remove string 'Size Charts' or 'size charts' from df['Brand']

```python
df['Brands'] = df['Brands'].str.replace('size charts', '', case=False, regex=True).str.strip()
df
```

## Make the first letter of the string from df['Brand'] capital. 

```python
df['Brands'] = df['Brands'].str.strip().str[0].str.upper() + df['Brands'].str.strip().str[1:]
df.head(40)
```

## Find the brands that DO NOT have XS or XL. 

```python
df.to_csv('updated_busts.csv')
```

```python
brands_with_xs = df.loc[df['Letter_sizing'] == 'XS', 'Brands'].unique()
brands_without_xs = df.loc[~df['Brands'].isin(brands_with_xs), 'Brands'].unique()
brands_without_xs
```

```python
brands_with_xl = df.loc[df['Letter_sizing'] == 'XL', 'Brands'].unique()
brands_without_xl = df.loc[~df['Brands'].isin(brands_with_xl), 'Brands'].unique()
brands_without_xl
```
