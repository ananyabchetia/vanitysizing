```python
import pandas as pd
df=pd.read_csv('tops.csv')
df
```

```python
df['Brands'].value_counts()
```

```python
df['Letter_sizing'].value_counts()
```

```python
## Find the smallest bust size for XS, S, M, L and XL 
filtered = df[df['Letter_sizing'].isin(['XS', 'S', 'M', 'L', 'XL'])]
result = filtered[filtered['Bust'] == filtered.groupby('Letter_sizing')['Bust'].transform('min')]
result
```

```python
## Find the largest bust size for XS, S, M, L and XL 
filtered = df[df['Letter_sizing'].isin(['XS', 'S', 'M', 'L', 'XL'])]
result = filtered[filtered['Bust'] == filtered.groupby('Letter_sizing')['Bust'].transform('max')]
result
```
